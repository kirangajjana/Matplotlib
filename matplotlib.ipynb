```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

PIE CHART

```python
name=["kiran gajjana","soham biswas","suman som","shubhani singh"]
working_hours=[20,10,9,8]
```

```python
plt.pie(working_hours,labels=name)
```

```python
plt.scatter(name,working_hours)
plt.xlabel("names")
plt.ylabel("working_hours")
plt.title("anual_working_hours")
```

```python
plt.plot(name,working_hours,color='red',marker="+",markersize=20)
plt.xlabel("names")
plt.ylabel("working_hours")
plt.title("total_working_hours")
```

```python
year=[2023,2024,2025,2026,2027]
income=[1000000,1500000,2000000,2500000,3000000]
```

```python
plt.bar(year,income,color='red')
plt.xlabel('year')
plt.ylabel('income')
plt.title("success in years")
```
